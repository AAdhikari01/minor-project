### Actual processing

```python
input_file = "../data/Separated_csv_1/non_native_human/all_human_filtered.csv"
output_folder = "../data/Separated_csv_1/non_native_human/long_text_chunks"

os.makedirs(output_folder, exist_ok=True)

chunk_size = 1_000_000  # 1M rows per chunk
chunk_id = 0

for chunk in pd.read_csv(input_file, chunksize=chunk_size):
    
    # Ensure text column is string & drop NaN
    chunk = chunk.dropna(subset=["text"])
    chunk["text"] = chunk["text"].astype(str)
    
    # Count words
    chunk["word_count"] = chunk["text"].str.split().apply(len)
    
    # Filter rows with > 50 words
    long_texts = chunk[chunk["word_count"] > 50].copy()
    
    # Drop helper column
    long_texts.drop(columns=["word_count"], inplace=True)
    
    # Save filtered chunk
    output_file = os.path.join(output_folder, f"long_text_chunk_{chunk_id}.csv")
    long_texts.to_csv(output_file, index=False, encoding="utf-8")
    
    print(f"Saved {len(long_texts)} rows → {output_file}")
    
    chunk_id += 1

print("Done processing all chunks!")
```

```python

input_file = "../data/Raw_data/responses.csv"
output_file = "../data/Separated_csv_1/responses_csv/responses_processed.csv"

# Read csv
df1 = pd.read_csv(input_file)

# Create new dataframe with only response_text
df_new = df1[["response_text"]].copy()

# Rename column to "text"
df_new.rename(columns={"response_text": "text"}, inplace=True)

# Add generated column with value 1
df_new["generated"] = 1

# Create folder if not exists
os.makedirs(os.path.dirname(output_file), exist_ok=True)

# Save to csv
df_new.to_csv(output_file, index=False, encoding="utf-8")

print("Saved:", output_file)
print(df_new.head())
```
