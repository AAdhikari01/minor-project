```python
import pandas as pd
import numpy as np
```

```python
human_df = pd.read_csv("human_text_only.csv")
ai_df = pd.read_csv("ai_text_only.csv")
```

```python
def split_dataframe(df,chunk_size):
    df = df.sample(frac=1,random_state=42).reset_index(drop=True)
    return [df.iloc[i:i+chunk_size]for i in range(0,len(df),chunk_size)]
human_chunks = split_dataframe(human_df,5000)
ai_chunks= split_dataframe(ai_df,5000)
```

```python
for idx,chunk in enumerate(human_chunks):
    chunk.to_csv(f'human_chunk_{idx+1}.csv',index=False)
for idx,chunk in enumerate(ai_chunks):
    chunk.to_csv(f'ai_chunk_{idx+1}.csv',index=False)
```
